```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
df= pd.read_csv('Swiggy_Orders.csv')
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.duplicated().sum()
```

```python
plt.figure(figsize=(8, 6))
sns.histplot(df['order_amount'], kde=True)
plt.title('Distribution of Order Amounts')
plt.xlabel('Order Amount')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.countplot(x='order_time', data=df)
plt.title('Number of Orders by Time Slot')
plt.xlabel('Time Slot')
plt.ylabel('Number of Orders')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(x='cuisine', data=df)
plt.title('Popularity of Cuisines')
plt.xlabel('Cuisine')
plt.ylabel('Number of Orders')
plt.show()
```

### Data Analysis Key Findings
* The majority of orders fall within the \$100 to \$300 price range, with the most frequent order amount being around \$200-\$250.
* The distribution of order amounts is right-skewed, suggesting a smaller number of high-value orders.
* "Dinner" is the busiest time slot for orders, followed by "Lunch," with "Breakfast" being the least busy.
* The dataset reveals distinct preferences for different types of cuisines, with some being ordered more frequently than others.

### Insights or Next Steps
* **Targeted Promotions:** Businesses can leverage these insights to offer targeted promotions. For instance, offering special deals during less busy hours like breakfast could help increase order volume. Similarly, promoting popular cuisines or creating combo deals around the average order amount could enhance sales.
* **Operational Adjustments:** Delivery services can optimize their operations by allocating more delivery partners during peak hours (Dinner and Lunch) to ensure timely deliveries and improve customer satisfaction.
