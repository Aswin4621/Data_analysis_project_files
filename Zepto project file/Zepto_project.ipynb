```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
avg_delivery_time = df.groupby('rider_id')['delivery_time_min'].mean().reset_index()
fig = px.bar(avg_delivery_time, x='rider_id', y='delivery_time_min', title='Average Delivery Time per Rider')
fig.show()
```

```python
fig = px.scatter_mapbox(df, lat="pickup_lat", lon="pickup_long",
                  color_continuous_scale=px.colors.cyclical.IceFire, size_max=15, zoom=10,
                  mapbox_style="carto-positron", title='Pickup Locations on Mapbox')
fig.show()
```

```python
fig = px.histogram(df, x='delivery_time_min', nbins= 30, title='Distribution of Delivery Times')
fig.show()
```
